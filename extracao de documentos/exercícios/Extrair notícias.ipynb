```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
# acessar a página de notícias da Anvisa
url = "https://www.gov.br/anvisa/pt-br/assuntos/noticias-anvisa"

html = requests.get(url).text
soup = BeautifulSoup(html, "html.parser")

dados = []
```

```python
# iterar sobre cada notícia encontrada na página
for noticia in soup.find_all("li"):

    titulo_tag = noticia.find("h2", class_="titulo")
    data_tag = noticia.find("span", class_="data")

    if titulo_tag:

        link_tag = titulo_tag.find("a")

        dados.append({
            "titulo": link_tag.get_text(strip=True),
            "link": link_tag["href"],
            "data": data_tag.get_text(strip=True) if data_tag else None
        })


#salvar em um dataframe do pandas
df = pd.DataFrame(dados)
```

```python
df
```

```python
#usar o link de cada notícia para acessar o conteúdo completo e extrair o texto
textos = []

for link in df["link"]:

    html = requests.get(link).text
    soup = BeautifulSoup(html, "html.parser")

    corpo = soup.select_one("#content-core")

    if corpo:
        paragrafos = corpo.find_all("p")
        texto = " ".join(p.get_text(strip=True) for p in paragrafos)
    else:
        texto = None

    textos.append(texto)

df["texto"] = textos

#salvar em um arquivo CSV
#df.to_csv("noticias_anvisa_com_texto.csv", index=False, encoding="utf-8-sig")
```

```python
df.to_csv("noticias_anvisa_com_texto.csv", index=False, encoding="utf-8-sig")
```
