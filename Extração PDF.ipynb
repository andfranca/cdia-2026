```python
#Importar o pymupdf
import pymupdf
```

```python
#Abrir o PDF e extrair o texto
caminho = "parecer.pdf"

doc = pymupdf.open(caminho)
#page = doc.load_page(0)

texto = ''

for pagina in doc:
    texto += pagina.get_text()
```

```python
print(texto)
```

```python
#para excluir o cabeçalho, é necessário definir uma área de recorte (clip) para a função get_text() do pymupdf.

caminho = 'PC_ Aeroporto Porto Alegre.pdf'
doc = pymupdf.open(caminho)

# Defina a altura do cabeçalho que deseja ignorar (em pontos/pixels)
ALTURA_CABECALHO = 80 

texto = ''

for pagina in doc:
    # Define a área útil: desconsidera o topo (y0) até a ALTURA_CABECALHO
    # page.rect retorna (x0, y0, x1, y1) -> (esquerda, topo, direita, base)
    area_util = pymupdf.Rect(pagina.rect.x0, ALTURA_CABECALHO, pagina.rect.x1, pagina.rect.y1)
    
    # Extrai o texto estritamente de dentro da área útil
    texto += pagina.get_text("text", clip=area_util)

doc.close()
```

```python
print(texto)
```

```python
# Salvar o texto extraído em um arquivo .txt
with open('texto_extraido.txt', 'w', encoding='utf-8') as f:
    f.write(texto) 
```
